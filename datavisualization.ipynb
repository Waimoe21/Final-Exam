```python
### Do not change this cell. 

import pandas as pd

df = pd.read_csv('Course Evaluation_Parami.csv')
df = df.drop('Timestamp', axis = 1)
df_students = df.iloc[:, 3:8]
stu_col_name = df_students.columns.to_list()
df_instructors = df.iloc[:, 8:13]
ins_col_name = df_instructors.columns.to_list()
df_course = df.iloc[:, 13:18]
course_col_name = df_course.columns.to_list()
df1 = df.drop(stu_col_name+ins_col_name+course_col_name, axis = 1)
```

```python
## Example to crate plot

import matplotlib.pyplot as plt

df1 = df1.sort_values(by = 'Age')
stu_age = df1['Age'].value_counts(normalize= True).to_frame()
plt.pie(stu_age['Age'], labels = stu_age.index, autopct = '%2.2f%%')
plt.show()
#comment : more than 50% are 20 years old. 
```

```python
# comment: female students attend the class regularly but male students have some outliers. 
# but on average, both perform pretty well at above 98%
female = df1[df1['Gender']=='Female']
male = df1[df1['Gender']=='Male']
plt.boxplot([female['Attendance %'], male['Attendance %']], labels = ['female', 'male'])
plt.show()
```

```python
recommend = df1['Would you recommend this course to your friend. '].value_counts().to_frame()
plt.bar(recommend.index, recommend['Would you recommend this course to your friend. '])
plt.show()
# comment 9 students strongly recommend. 
```

```python
reason = df1['Why did you choose this course?'].value_counts(normalize= True).to_frame()
plt.pie(reason['Why did you choose this course?'], labels = reason.index, autopct = '%2.2f%%')
plt.show()
```
